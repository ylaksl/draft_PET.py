```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def fresnel_integrals(t):
    C = np.array([quad(lambda u: np.cos(np.pi/2 * u**2), 0, ti)[0] for ti in t])
    S = np.array([quad(lambda u: np.sin(np.pi/2 * u**2), 0, ti)[0] for ti in t])
    return C, S

def plot_euler_spiral(t_max=5, num_points=1000):
    # Generate parameter values
    t = np.linspace(-t_max, t_max, num_points)
    
    # Compute Fresnel integrals manually
    x, y = fresnel_integrals(t)
    
    # Plot the spiral
    plt.figure(figsize=(8, 8))
    plt.plot(x, y, label="Euler Spiral")
    plt.xlabel("Fresnel C(t)")
    plt.ylabel("Fresnel S(t)")
    plt.title("Euler (Cornu) Spiral")
    plt.axhline(0, color='gray', linestyle='--')
    plt.axvline(0, color='gray', linestyle='--')
    plt.legend()
    plt.show()

# Call the function to plot
t_max = 5  # Adjust the maximum parameter value for the plot
plot_euler_spiral(t_max=t_max)
```

```python
t = np.linspace(-5,5,100)
c, s = fresnel_integrals(t)

plt.plot(t, c, label = '$C(t)$')
plt.plot(t, s, label = '$S(t)$')
plt.legend()
```

```python
def diffraction_pattern():
    wavelength = 5e-8
    k = (2 * np.pi) / wavelength
    xi0 = 1
    side = 1e-6
    points = 500
    spacing = side / points
    
    x1 = side / 2
    y1 = side / 2
    
    xi = np.empty([points, points], float)
    
    for a in range(points):
        y = spacing * a
        for j in range(points):
            x = spacing * j
            r = np.sqrt((x - x1) ** 2 + (y - y1) ** 2)
            c, s = fresnel_integrals([r])
            xi[a, j] = xi0 * c[0]
    
    plt.figure(figsize=(8, 8))
    plt.imshow(xi, cmap='gray', vmax=0.1, extent=[0, side, 0, side])
    plt.title("Fresnel Diffraction Pattern")
    plt.xlabel("x (m)")
    plt.ylabel("y (m)")
    plt.colorbar(label="Intensity")
    plt.show()
```
