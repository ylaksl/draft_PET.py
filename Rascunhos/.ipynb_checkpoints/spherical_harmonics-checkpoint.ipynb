```python
import numpy as np
import scipy.special as sp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def spherical_harmonic(l, m, theta, phi):
    """
    Compute the spherical harmonic Y_l^m(theta, phi) for given l, m, theta, and phi.
    """
    # Compute the associated Legendre polynomial P_l^m(cos(theta))
    P_lm = sp.lpmv(m, l, np.cos(theta))  # Associated Legendre polynomial P_l^m(cos(theta))
    
    # Compute the normalization factor
    norm = np.sqrt((2*l + 1) / (4 * np.pi) * np.math.factorial(l - m) / np.math.factorial(l + m))
    
    # Compute the spherical harmonic
    Y_lm = norm * P_lm * np.exp(1j * m * phi)  # Complex exponential e^(im*phi)
    
    return Y_lm

# Generate a grid of theta and phi
theta = np.linspace(0, np.pi, 100)  # Colatitude from 0 to pi
phi = np.linspace(0, 2*np.pi, 100)  # Azimuthal angle from 0 to 2pi
theta, phi = np.meshgrid(theta, phi)  # Create meshgrid

# Choose l and m
l = 3
m = 2

# Compute the spherical harmonic on the grid
Y_lm = spherical_harmonic(l, m, theta, phi)

# Extract the real part of the spherical harmonic for visualization
Y_real = np.real(Y_lm)

# Convert spherical coordinates (theta, phi) to Cartesian coordinates for 3D plotting
x = Y_real * np.sin(theta) * np.cos(phi)
y = Y_real * np.sin(theta) * np.sin(phi)
z = Y_real * np.cos(theta)

# Plotting the spherical harmonic on a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, rstride=1, cstride=1, facecolors=plt.cm.viridis((Y_real - Y_real.min()) / (Y_real.max() - Y_real.min())), 
                linewidth=0, antialiased=False, shade=True)

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(f"Spherical Harmonic Y_{l}^{m} (Real part)")

plt.show()
```
