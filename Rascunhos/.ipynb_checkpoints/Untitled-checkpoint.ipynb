```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.patches import Circle

# Double Pendulum Parameters
m1, m2 = 1.0, 1.0  # masses
l1, l2 = 1.0, 1.0  # lengths
g = 9.81           # gravity

def double_pendulum(t, y):
    """Equations of motion for the double pendulum."""
    θ1, ω1, θ2, ω2 = y
    
    Δθ = θ2 - θ1
    den1 = (m1 + m2)*l1 - m2*l1*np.cos(Δθ)**2
    den2 = (l2/l1)*den1
    
    dθ1dt = ω1
    dω1dt = (m2*l1*ω1**2*np.sin(Δθ)*np.cos(Δθ) + m2*g*np.sin(θ2)*np.cos(Δθ) + 
              m2*l2*ω2**2*np.sin(Δθ) - (m1 + m2)*g*np.sin(θ1)) / den1
    
    dθ2dt = ω2
    dω2dt = (-m2*l2*ω2**2*np.sin(Δθ)*np.cos(Δθ) + (m1 + m2)*g*np.sin(θ1)*np.cos(Δθ) - 
              (m1 + m2)*l1*ω1**2*np.sin(Δθ) - (m1 + m2)*g*np.sin(θ2)) / den2
    
    return np.array([dθ1dt, dω1dt, dθ2dt, dω2dt])

def compute_lyapunov_exponents(initial_conditions, t_span=(0, 50), dt=0.01):
    """Compute the largest Lyapunov exponent (LLE) for given initial conditions."""
    n = 4  # Phase space dimension
    Q = np.eye(n)  # Orthonormal basis for tangent space
    λ_sum = np.zeros(n)
    t_eval = np.arange(t_span[0], t_span[1], dt)
    
    # Reference trajectory
    sol_ref = solve_ivp(double_pendulum, t_span, initial_conditions, t_eval=t_eval)
    
    # Perturbations (tangent space)
    y0_tangent = np.concatenate((initial_conditions, Q.flatten()))
    
    # Integrate variational equations (approximate Jacobian)
    def variational_eq(t, y):
        θ1, ω1, θ2, ω2 = sol_ref.y[:, np.argmin(np.abs(sol_ref.t - t))]
        δy = y[4:].reshape(n, n)
        
        # Numerical Jacobian
        eps = 1e-8
        J = np.zeros((n, n))
        for i in range(n):
            y_plus = np.array([θ1, ω1, θ2, ω2])
            y_plus[i] += eps
            f_plus = double_pendulum(t, y_plus)
            y_minus = np.array([θ1, ω1, θ2, ω2])
            y_minus[i] -= eps
            f_minus = double_pendulum(t, y_minus)
            J[:, i] = (f_plus - f_minus) / (2 * eps)
        
        dδy_dt = (J @ δy).flatten()
        return np.concatenate((double_pendulum(t, [θ1, ω1, θ2, ω2]), dδy_dt))
    
    sol_tangent = solve_ivp(variational_eq, t_span, y0_tangent, t_eval=t_eval)
    
    # Reorthogonalize and compute exponents
    for i in range(1, len(t_eval)):
        δy = sol_tangent.y[4:, i].reshape(n, n)
        Q, R = np.linalg.qr(δy)
        λ_sum += np.log(np.abs(np.diag(R)))
    
    λ = λ_sum / (t_span[1] - t_span[0])
    return λ

# ===== Part 1: Compute LLE for varying initial θ1, θ2 =====
θ1_range = np.linspace(0, np.pi, 20)  # Initial θ1 values
θ2_range = np.linspace(0, np.pi, 20)  # Initial θ2 values
LLE_grid = np.zeros((len(θ1_range), len(θ2_range)))

for i, θ1 in enumerate(θ1_range):
    for j, θ2 in enumerate(θ2_range):
        λ = compute_lyapunov_exponents([θ1, 0, θ2, 0], t_span=(0, 20))
        LLE_grid[i, j] = λ[0]  # Largest Lyapunov exponent

# Plot LLE heatmap
plt.figure(figsize=(10, 8))
plt.imshow(LLE_grid, extent=[0, np.pi, 0, np.pi], origin='lower', cmap='viridis')
plt.colorbar(label='Largest Lyapunov Exponent (LLE)')
plt.xlabel('Initial θ2 (rad)')
plt.ylabel('Initial θ1 (rad)')
plt.title('Chaotic Regions of Double Pendulum')
plt.show()

# ===== Part 2: Animate chaotic vs. regular motion =====
def animate_pendulum(initial_conditions, t_span=(0, 10), dt=0.05):
    """Animate the double pendulum for given initial conditions."""
    t_eval = np.arange(t_span[0], t_span[1], dt)
    sol = solve_ivp(double_pendulum, t_span, initial_conditions, t_eval=t_eval)
    θ1, θ2 = sol.y[0], sol.y[2]
    
    # Convert angles to Cartesian coordinates
    x1 = l1 * np.sin(θ1)
    y1 = -l1 * np.cos(θ1)
    x2 = x1 + l2 * np.sin(θ2)
    y2 = y1 - l2 * np.cos(θ2)
    
    # Set up animation
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.set_xlim(-2.2, 2.2)
    ax.set_ylim(-2.2, 2.2)
    ax.set_aspect('equal')
    ax.grid()
    
    line, = ax.plot([], [], 'o-', lw=2)
    time_text = ax.text(0.02, 0.95, '', transform=ax.transAxes)
    
    def init():
        line.set_data([], [])
        time_text.set_text('')
        return line, time_text
    
    def update(frame):
        line.set_data([0, x1[frame], x2[frame]], [0, y1[frame], y2[frame]])
        time_text.set_text(f'Time = {t_eval[frame]:.2f}s')
        return line, time_text
    
    ani = FuncAnimation(fig, update, frames=len(t_eval), init_func=init, blit=True, interval=50)
    plt.close()
    return ani

# Example animations
print("Animating chaotic motion (θ1=π/2, θ2=π/2)...")
ani_chaotic = animate_pendulum([np.pi/2, 0, np.pi/2, 0])
ani_chaotic.save('chaotic_pendulum.gif', writer='pillow', fps=20)

print("Animating regular motion (θ1=π/4, θ2=π/4)...")
ani_regular = animate_pendulum([np.pi/4, 0, np.pi/4, 0])
ani_regular.save('regular_pendulum.gif', writer='pillow', fps=20)
```
