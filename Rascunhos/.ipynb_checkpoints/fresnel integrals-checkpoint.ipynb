```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def fresnel_integrals(t):
    C = np.array([quad(lambda u: np.cos(np.pi/2 * u**2), 0, ti)[0] for ti in t])
    S = np.array([quad(lambda u: np.sin(np.pi/2 * u**2), 0, ti)[0] for ti in t])
    return C, S

def plot_euler_spiral(t_max=5, num_points=1000):
    # Generate parameter values
    t = np.linspace(-t_max, t_max, num_points)
    
    # Compute Fresnel integrals manually
    x, y = fresnel_integrals(t)
    
    # Plot the spiral
    plt.figure(figsize=(8, 8))
    plt.plot(x, y, label="Euler Spiral")
    plt.xlabel("Fresnel C(t)")
    plt.ylabel("Fresnel S(t)")
    plt.title("Euler (Cornu) Spiral")
    plt.axhline(0, color='gray', linestyle='--')
    plt.axvline(0, color='gray', linestyle='--')
    plt.legend()
    plt.show()

# Call the function to plot
t_max = 5  # Adjust the maximum parameter value for the plot
plot_euler_spiral(t_max=t_max)
```

```python
t = np.linspace(-5,5,100)
c, s = fresnel_integrals(t)

plt.plot(t, c, label = '$C(t)$')
plt.plot(t, s, label = '$S(t)$')
plt.legend()
```

```python
def diffraction_pattern():
    wavelength = 5e-8
    k = (2 * np.pi) / wavelength
    xi0 = 1
    side = 1e-6
    points = 500
    spacing = side / points
    
    x1 = side / 2
    y1 = side / 2
    
    xi = np.empty([points, points], float)
    
    for a in range(points):
        y = spacing * a
        for j in range(points):
            x = spacing * j
            r = np.sqrt((x - x1) ** 2 + (y - y1) ** 2)
            c, s = fresnel_integrals([r])
            xi[a, j] = xi0 * c
        print(f'c = {c[0]}')
    
    plt.figure(figsize=(8, 8))
    plt.imshow(xi, cmap='gray', vmax=0.1, extent=[0, side, 0, side])
    plt.title("Fresnel Diffraction Pattern")
    plt.xlabel("x (m)")
    plt.ylabel("y (m)")
    plt.colorbar(label="Intensity")
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import fresnel

# Define the complex Fresnel S(z) integral (using scipy's fresnel for real parts)
def complex_fresnel_S(z):
    # For complex z, separate into real and imaginary parts
    # (Note: scipy's fresnel only handles real inputs, so we approximate here)
    # For a proper implementation, use numerical integration for complex z.
    S_real, C_real = fresnel(z.real)
    S_imag, C_imag = fresnel(z.imag)
    return S_real + 1j * S_imag  # Simplified approximation

# Create a grid of complex points
x = np.linspace(-2, 2, 500)
y = np.linspace(-2, 2, 500)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Compute S(z) for each point
S_values = np.vectorize(complex_fresnel_S)(Z)

# Plot the magnitude |S(z)|
plt.figure(figsize=(10, 8))
plt.pcolormesh(X, Y, np.abs(S_values), shading='auto', cmap='viridis')
plt.colorbar(label='Magnitude |S(z)|')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.title('Fresnel Integral S(z) (Magnitude)')
plt.show()

# Plot the phase (argument) of S(z)
plt.figure(figsize=(10, 8))
plt.pcolormesh(X, Y, np.angle(S_values), shading='auto', cmap='hsv')
plt.colorbar(label='Phase (radians)')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.title('Fresnel Integral S(z) (Phase)')
plt.show()
```
